**Aula 2 - Importação**

```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('/content/Imersão Python - Tabela de ações.xlsx', sheet_name='Principal')
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel('/content/Imersão Python - Tabela de ações.xlsx', sheet_name='Total_de_acoes')
df_total_acoes.head()
```

```python
df_ticker = pd.read_excel('/content/Imersão Python - Tabela de ações.xlsx', sheet_name='Ticker')
df_ticker.head()
```

```python
df_chatgpt = pd.read_excel('/content/Imersão Python - Tabela de ações.xlsx', sheet_name='ChatGPT')
df_chatgpt.head()
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Data': 'data', 'Ativo': 'ativo', 'Último (R$)': 'valor_final', 'Var. Dia (%)': 'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'qtd_teorica'}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal = df_principal.rename(columns={'Idade (em anos)': 'idade'})
df_principal
```

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
maior = df_principal['variacao'].max()
menor = df_principal['variacao'].min()
media = df_principal['variacao'].mean()

media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao'].mean()

print(f'Maior\tR$ {maior:,.2f}')
print(f'Menor\tR$ {menor:,.2f}')
print(f'Média\tR$ {media:,.2f}')
print(f'Média de quem subiu\tR$ {media_subiu:,.2f}')
print(f'Média de quem desceu\tR$ {media_desceu:,.2f}')
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao', text='variacao', title='Variação por Resultado (R$)')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, values='variacao', names='Segmento', title='Variation by Segment')
fig.show()
```
